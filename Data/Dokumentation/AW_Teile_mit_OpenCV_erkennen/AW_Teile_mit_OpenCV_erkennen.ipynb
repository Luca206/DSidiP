```python
from matplotlib.pyplot import figure
from matplotlib import pyplot as plt
import cv2 as cv
import numpy as np
import os
```

```python
input_filepath = "Eingabe_Bilder/"
output_filepath = "Ausgabe_Bilder/"
files = os.listdir(input_filepath)
filenames = [f"0{x}" for x in range(1, len(files) + 1)]
lower_red = np.array([  5,   0,   0])
upper_red = np.array([355, 255, 255])

for filename in filenames:
    image = cv.imread(filename = input_filepath +
                                 filename +
                                 ".jpg")
    hsv = cv.cvtColor(src = image,
                      code = cv.COLOR_BGR2HSV)
    mask = cv.inRange(src = hsv,
                      lowerb = lower_red,
                      upperb = upper_red)
    figure(figsize = (10.24, 7.68),
           dpi = 100)
    plt.imshow(X = mask,
               cmap = "binary")
    plt.title(label = f"{output_filepath +
                         filename +
                         ".png"} (Lichtquelle: unten rechts)")
    plt.tick_params(left = False,
                    bottom = False,
                    labelleft = False,
                    labelbottom = False)
    plt.colorbar()
    plt.savefig(fname = output_filepath +
                        filename +
                        ".png")
    plt.show()
```
